```python
import matplotlib.pyplot as plt       #importing tool for plotting

def dist_btw_two_points(p1,p2):       #This function takes two points as input and returns the distance between them rounded to the fourth decimal
    distance = (( ((p2[0]-p1[0])**2) + ((p2[1]-p1[1])**2) )**0.5)
    distance = round(distance*10000)
    distance /= 10000
    return distance

def sort(point):                    #The purpose of this function is to sort the points obtained from user i increasing order of their distance
    for i in range(1,len(point)):
        if dist_btw_two_points(point[i-1],[0,0]) > dist_btw_two_points(point[i],[0,0]):
            temp = point[i-1]
            point[i-1] = point[i]
            point[i] = temp
            sort(point)

def point_array():                  #The purpose of this function is to take input of points from the user and return a sorted array of destinations
    n = int(input('Enter the number of destinations: '))
    destination_point = []
    sorted_point = []
    for i in range(n):
        p = []
        x = float(input(f"Enter the x co-ordinate of point {i+1} respectively: "))
        p.append(x)
        y = float(input(f"Enter the y co-ordinate of point {i+1} respectively: "))
        p.append(y)
        destination_point.append(p)
    sort(destination_point)
    return destination_point

def constraint_table(point):         #This function takes input the sorted destinations and returns the constraint table used in our algorithm in the form of a 2D matrix
    table = []                      
    for i in range(len(point)):
        rows = []
        for j in range(len(point)):
            dist  = dist_btw_two_points(point[j],[0,0])-dist_btw_two_points(point[j],point[i])
            rows.append(dist)
        table.append(rows)
    return table    

def plot_point(array_point):       #This function takes the sorted destinations as input and plots them on the graph
    plot1 = plt.figure(1)
    arr = []                       
    plt.plot([0],[0],'rs')         #Starting and ending points of all drones is same and represented by red
    for i in range(len(array_point)):
        x_coord = []
        y_coord = []
        x_coord.append(array_point[i][0])
        y_coord.append(array_point[i][1])
        arr.append(x_coord)
        arr.append(y_coord)
        arr.append('bo')
        plt.plot(x_coord,y_coord,'bo') #Destinations is represented by blue circle
    plt.title("Destinations")
    plot2 = plt.figure(2)
    arr = []                       
    plt.plot([0],[0],'rs')         #Starting and ending points of all drones is same and represented by red
    for i in range(len(array_point)):
        x_coord = []
        y_coord = []
        x_coord.append(array_point[i][0])
        y_coord.append(array_point[i][1])
        arr.append(x_coord)
        arr.append(y_coord)
        arr.append('bo')
        plt.plot(x_coord,y_coord,'bo') #Destinations is represented by blue circle

def plot_line(array_point):    #This function takes an array of destination as input and plots a line starting from the first point and joining it to the next element in the array and finaaly ending at the starting point
    plot2 = plt.figure(2)
    arr = []                   
    plt.plot([0],[0],'rs')
    x_coord = [0]
    y_coord = [0]
    for i in range(len(array_point)):
        x_coord.append(array_point[i][0])
        y_coord.append(array_point[i][1])
    x_coord.append(0)
    y_coord.append(0)
    plt.plot(x_coord,y_coord,'b-')
    plt.title("Paths")



def possible_paths_of_a_point(dt_cpy,length,temp_array): #This function creates the search tree for the point which is farthest from the origin.
    r = 3                                                #i.e. It creates the search tree of all possible paths along with total distance travelled for the last point in dt_cpy and stores it in temp_array.
    table = constraint_table(dt_cpy)
    anv1 = True
    for j in range(length):
        if table[j][length-1]>=0 and j!=length-1: 
            anv1 = False
            anv2 = True
            for k in range(length):
                if table[k][j]>=0 and k!=length-1 and k!=j: 
                    anv2 = False
                    if anv1 == False and anv2 == False:
                        temp_array.append([[dt_cpy[k],dt_cpy[j],dt_cpy[length-1]],[dist_btw_two_points(dt_cpy[j],dt_cpy[length-1])+dist_btw_two_points(dt_cpy[j],dt_cpy[k])+dist_btw_two_points(dt_cpy[k],[0,0])]])
                        r = 2
                if anv2 == True and anv1 == False:
                    temp_array.append([[dt_cpy[j],dt_cpy[length-1]],[dist_btw_two_points(dt_cpy[j],dt_cpy[length-1])+dist_btw_two_points(dt_cpy[j],[0,0])]])
                    r = 1
        if anv1 == True:
            temp_array.append([[dt_cpy[length-1]],[dist_btw_two_points(dt_cpy[length-1],[0,0])]])
            r = 0
    #for k in range(len(temp_array)):
        #print(temp_array[k])
    return r
#P1,P2,P3,P4,P5
#[[[P1,P2,P5],[D1]],[[P4,P3,P5],[D2]]], if D1<D2, d =D1, s = 0
path = []
def deciding_path():     #This is the main brain of the program. It calls the possible_paths_of_a_point in an orderly manner and choses the path in whic maximum number of destinations along with the least distance is travelled.
    global s             #Then it appends that specific trajectory to the array named path
    destinations = point_array()
    dt_cpy = destinations.copy()
    print(f"Destinations are: {destinations}")
    print("")
    print("Constraint table: ")
    table = constraint_table(dt_cpy)
    for i in range(len(dt_cpy)):
        print(table[i])
    while True:
        table = constraint_table(dt_cpy)
        if len(dt_cpy) == 0:
            break
        if len(dt_cpy) == 1:
            path.append([[dt_cpy[0]],[dist_btw_two_points(dt_cpy[0],[0,0])]])
            dt_cpy.remove(dt_cpy[0])
        if len(dt_cpy)!=0:
            temp_array = []
            if possible_paths_of_a_point(dt_cpy,len(dt_cpy),temp_array) == 2:
                d = float('inf')
                for t in range(len(temp_array)):
                    if temp_array[t][1][0] < d and len(temp_array[t][0]) == 3:
                        d = temp_array[t][1][0]
                        s = t
                path.append(temp_array[s])
                dt_cpy.remove(temp_array[s][0][0])
                dt_cpy.remove(temp_array[s][0][1])
                dt_cpy.remove(temp_array[s][0][2])
            if possible_paths_of_a_point(dt_cpy,len(dt_cpy),temp_array) == 1:
                d = float('inf')
                for t in range(len(temp_array)):
                    if temp_array[t][1][0] < d and len(temp_array[t][0]) == 2:
                        d = temp_array[t][1][0]
                        s = t
                path.append(temp_array[s])
                dt_cpy.remove(temp_array[s][0][0])
                dt_cpy.remove(temp_array[s][0][1])
            if possible_paths_of_a_point(dt_cpy,len(dt_cpy),temp_array) == 0:
                d = float('inf')
                for t in range(len(temp_array)):
                    if temp_array[t][1][0] < d and len(temp_array[t][0]) == 1:
                        d = temp_array[t][1][0]
                        s = t
                path.append(temp_array[s])
                dt_cpy.remove(temp_array[s][0][0])
    return destinations

destinations = deciding_path()
print("")
print("Paths are:")
for i in range(len(path)):
    print(path[i])

plot_point(destinations) #plotting destinations
for i in range(len(path)):#plotting trajectory of drones
    plot_line(path[i][0])
```
